```python
from sage.all import *
# Sin coeffs
def format_hex(value):
    l = hex(value)[2:]
    n = 8
    x = [l[i:i + n] for i in range(0, len(l), n)]
    return "0x" + "_".join(x) + "_u128"

def print_dyadic(value):
    (s, m, e) = RealField(128)(value).sign_mantissa_exponent();
    print("DyadicFloat128 {")
    print(f"    sign: DyadicSign::{'Pos' if s >= 0 else 'Neg'},")
    print(f"    exponent: {e},")
    print(f"    mantissa: {format_hex(m)},")
    print("},")
    
arr = [1,
-0.49991338831735860796075102864431055957823229889929,
0.33238337709413673145411728133516329485319077195155,
-0.245127343744325628527317048682869543952450462234626,
0.184412768602682174251580612983506889721286234089187,
-0.131586006397736043891904341738580411124979713180765,
8.317561792389009332309294250494051612441049439579e-2,
-4.3896515214761936546723431427559033709609352506573e-2,
1.83614314235049107959483571402956924578584275711765e-2,
-5.7700807074953497849976259252866523930875610860376e-3,
1.26999703967516546474634122645358915431481344808102e-3,
-1.7377162990984217050209600145934969736767673853033e-4,
1.1094487643661973850983779415945105909504148056633e-5]

for num in arr:
    print_dyadic(num)
```
